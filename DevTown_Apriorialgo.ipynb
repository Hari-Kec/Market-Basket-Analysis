```python
df['Itemname'] = df['Itemname'].str.strip() 
df.dropna(axis=0, subset=['BillNo'], inplace=True) 
df['BillNo'] = df['BillNo'].astype('str') 
df = df[~df['BillNo'].str.contains('C')]
```

```python
df.shape
```

```python
basket = (df[df['Country'] =="France"] 
          .groupby(['BillNo', 'Itemname'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('BillNo')) 
```

```python
basket.head()
```

```python
basket.shape
```

```python
def result(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_set = basket.applymap(result)
```

```python
basket_set.head()
```

```python
 frequent_itemsets = apriori(basket_set.astype('bool'), min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[(rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8)]
```
